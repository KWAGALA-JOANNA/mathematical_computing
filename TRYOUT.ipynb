```python
# Example
# given 10 trials for coin toss to generate 10 data points:
from numpy import random
x=random.binomial(n=10, p=0.5, size=10)
x
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)
plt.show()
```

```python
#Generate a random distribution of size 2x3:
y =random.normal(size=(2,3))
y
```

```python
#Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:
x = random.normal(loc=1, scale=2, size=(2,3))
x
```

```python
sns.distplot(random.normal(size=1000), hist=False)
```

```python
sns.distplot(random.normal(size=1000), hist=True)
```

```python
# difference between normal and binomial
sns.distplot(random.normal(loc=1, scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=False, label='binomial')
plt.legend()
```

```python
# Generate a random 1x10 distribution for occurrences 2
z = random.poisson(lam=2, size=10)
z
```

```python
sns.distplot(random.poisson(lam=2,size=1000), kde=False)
```
